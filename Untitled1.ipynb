```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data into a pandas DataFrame
url = 'visits.csv'
df = pd.read_csv(url)

# set the 'Time' column as the index
df.set_index('Time', inplace=True)

# plot the time series for all visits
plt.plot(df['All visits'])

# set title and axis labels
plt.title('Website visits over time')
plt.xlabel('Time')
plt.ylabel('Number of visits')

# compute and plot the rolling average
window = 12  # 12 months (1 year)
rolling_avg = np.convolve(df['All visits'], np.ones(window), 'valid') / window
plt.plot(df.index[window-1:], rolling_avg, color='red', linestyle='--')

# show the plot
plt.show()
```
